```python
import pandas as pd
import geopandas as gpd
```

```python
# Location of data sources
polling_places_path = "../Data/GeneralPollingPlacesDownload.csv"
tpp_polling_path = "../Data/HouseTppByPollingPlaceDownload.csv"
```

```python
def load_csv_data(data_path, skip_rows = 0):
    data = pd.read_csv(data_path, skiprows = skip_rows)
    lowercase = lambda x: str(x).lower()
    data.rename(lowercase, axis='columns', inplace=True)
    return data
```

```python
polling_places = load_csv_data(polling_places_path, 1)
polling_places.head()
```

```python
tpp_polling = load_csv_data(tpp_polling_path, 1)
tpp_polling.head()
```

```python
merged_polling = pd.merge(left = polling_places, right = tpp_polling, how = "left", left_on = "pollingplaceid", right_on = "pollingplaceid")
```

```python
merged_polling.head()
```

```python
merged_polling_places = merged_polling.loc[:,["state","divisionnm_x","pollingplaceid","pollingplacenm","premisesnm"
                      ,"premisessuburb","premisespostcode","latitude","longitude"
                      ,"liberal/national coalition votes","liberal/national coalition percentage"
                      ,"australian labor party votes","australian labor party percentage","totalvotes"]]
```

```python
merged_polling_places.head()
```

```python
g_sa2_file = gpd.read_file("../Data/sa2_2016_aust_shape/SA2_2016_AUST.shp")
```

```python
g_sa2_file.head()
```

```python
#Create geopandas dataframe with lat/long from polling places
g_polling_places = gpd.GeoDataFrame(merged_polling_places,
                                    geometry = gpd.points_from_xy(merged_polling_places.longitude,
                                                                  merged_polling_places.latitude),
                                   crs = "epsg:4326")
```

```python
g_polling_places.head()
```

```python
g_polling_places.crs
```

```python
g_sa2_file.crs = "epsg:4326"
```

```python
g_polling_places = g_polling_places.loc[g_polling_places.is_valid]
g_sa2_file = g_sa2_file.loc[g_sa2_file.is_valid]
```

```python
g_combined_df = gpd.sjoin(g_polling_places, g_sa2_file, how = "left", op = "intersects")
```

```python
g_combined_df.drop('geometry',axis=1).to_csv(r'../Data/polling_places_sa2.csv') 
```

```python
# Not get cumulatives for each SA2
combined_df = pd.DataFrame(g_combined_df.drop(columns = "geometry"))
agg_df = combined_df.groupby(["SA2_MAIN16", "state"])["liberal/national coalition votes","australian labor party votes","totalvotes"].sum()
agg_df.to_csv(r'../Data/sa2_votes.csv')
```
